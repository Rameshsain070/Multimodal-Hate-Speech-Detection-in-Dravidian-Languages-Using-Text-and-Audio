```python
# ==============================================================================
# CELL: MULTI-MODEL AUDIO TRAINING (TELUGU - 80/10/10 SPLIT)
# ==============================================================================

import os
import gc
import shutil
import numpy as np
import pandas as pd
import torch
import librosa
import evaluate
from torch import nn
from sklearn.model_selection import train_test_split
from sklearn.utils.class_weight import compute_class_weight
from datasets import Dataset
from audiomentations import Compose, AddGaussianNoise, TimeStretch, PitchShift
from transformers import (
    AutoModelForAudioClassification,
    AutoFeatureExtractor,
    TrainingArguments,
    Trainer,
    default_data_collator
)
from huggingface_hub import login

# ==========================================
# 1. CONFIGURATION (TELUGU)
# ==========================================
class Config:
    # --- Telugu Paths ---
    AUDIO_DIR = "/kaggle/input/hate-speech-detection-in-dravidian-languages/Hate Speech Detection in Dravidian languages/train/telugu/audio"
    CSV_PATH = "/kaggle/input/hate-speech-detection-in-dravidian-languages/Hate Speech Detection in Dravidian languages/train/telugu/text/TE-AT-train.csv"
    
    # --- Telugu-specific models (UPDATED) ---
    MODELS_TO_TRAIN = {
        # 1. Telugu Specific Wav2Vec2 (Switched to Anuragshas - Most stable/popular)
        "wav2vec2": "anuragshas/wav2vec2-large-xlsr-53-telugu", 
        
        # 2. WavLM (Best English base, performs well on Dravidian)
        "wavlm": "microsoft/wavlm-base",
        
        # 3. MMS (Massively Multilingual - Includes Telugu support)
        "mms": "facebook/mms-300m" 
    }
    
    NUM_CLASSES = 2  
    # Label Map (0 = Non-Hate, 1 = Hate)
    LABEL_MAP = {'N': 0, 'C': 1, 'G': 1, 'P': 1, 'R': 1, 'O': 1} 
    
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
    
    # Params
    MAX_DURATION = 5.0
    SAMPLING_RATE = 16000
    BATCH_SIZE = 4
    GRAD_ACCUMULATION = 4
    LEARNING_RATE = 3e-5
    EPOCHS = 20 

    # HF Info
    HF_USERNAME = "ramesh070"
    HF_TOKEN = "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" 

# Login to HF
login(token=Config.HF_TOKEN)

# ==========================================
# 2. DATA PREPARATION (COMMON SPLIT)
# ==========================================
print("ðŸ”µ 1. Loading and Splitting Data (Telugu)...")

if not os.path.exists(Config.AUDIO_DIR):
    raise FileNotFoundError(f"âŒ Audio Directory not found at: {Config.AUDIO_DIR}")

try:
    df = pd.read_csv(Config.CSV_PATH)
    print(f"   ðŸ” Found Columns: {df.columns.tolist()}") 
    
    # Standardize Text/Label Columns if needed
    if 'Transcript' in df.columns:
        df = df.rename(columns={'Transcript': 'text'})
    if 'Class Label Short' in df.columns:
        df = df.rename(columns={'Class Label Short': 'label_str'})
        
except Exception as e:
    print(f"âŒ Error loading CSV: {e}")
    raise e

# Map Labels
df['label'] = df['label_str'].map(Config.LABEL_MAP)

# --- FIX: Dynamic Audio Column Detection ---
possible_audio_cols = ['File_Name', 'File Name', 'filename', 'file_name', 'audio_id', 'id']
audio_col = None

for col in possible_audio_cols:
    if col in df.columns:
        audio_col = col
        break

if audio_col is None:
    raise KeyError(f"âŒ Could not find audio filename column. Available columns: {df.columns.tolist()}")

print(f"   âœ… Using '{audio_col}' as audio filename column.")

def get_audio_path(filename):
    filename = str(filename).strip()
    if not filename.lower().endswith('.wav'):
        filename += '.wav'
    return os.path.join(Config.AUDIO_DIR, filename)

df['audio_path'] = df[audio_col].apply(get_audio_path)
df = df[df['audio_path'].apply(os.path.exists)]
df = df.dropna(subset=['label'])
df['label'] = df['label'].astype(int)

# --- STATIC SPLIT LOGIC (80/10/10) ---
train_df, temp_df = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)
eval_df, test_df = train_test_split(temp_df, test_size=0.5, stratify=temp_df['label'], random_state=42)

# --- SAVE TEST SPLIT ---
test_save_path = "telugu_audio_test_split.csv"
test_df.to_csv(test_save_path, index=False)
print(f"ðŸ’¾ SUCCESS: Audio Test split saved to '{test_save_path}'")

print(f"âœ… Data Splits Created:")
print(f"   Train: {len(train_df)} | Val: {len(eval_df)} | Test: {len(test_df)}")

# Calculate Weights
class_weights_arr = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(train_df['label']),
    y=train_df['label']
)
class_weights = torch.tensor(class_weights_arr, dtype=torch.float).to(Config.DEVICE)
print(f"âš–ï¸ Class Weights: {class_weights.cpu().numpy()}")

# ==========================================
# 3. TRAINING LOOP (ALL 3 MODELS)
# ==========================================

augmentations = Compose([
    AddGaussianNoise(min_amplitude=0.001, max_amplitude=0.015, p=0.5),
    TimeStretch(min_rate=0.8, max_rate=1.25, p=0.5),
    PitchShift(min_semitones=-4, max_semitones=4, p=0.5),
])

class WeightedTrainer(Trainer):
    def compute_loss(self, model, inputs, return_outputs=False, **kwargs):
        labels = inputs.get("labels")
        outputs = model(**inputs)
        logits = outputs.get("logits")
        loss_fct = nn.CrossEntropyLoss(weight=class_weights)
        loss = loss_fct(logits.view(-1, self.model.config.num_labels), labels.view(-1))
        return (loss, outputs) if return_outputs else loss

def compute_metrics(eval_pred):
    f1_metric = evaluate.load("f1")
    acc_metric = evaluate.load("accuracy")
    logits, labels = eval_pred
    preds = np.argmax(logits, axis=-1)
    f1 = f1_metric.compute(predictions=preds, references=labels, average="binary")["f1"]
    acc = acc_metric.compute(predictions=preds, references=labels)["accuracy"]
    return {"accuracy": acc, "f1": f1}

# --- START LOOP ---
for short_name, model_id in Config.MODELS_TO_TRAIN.items():
    print("\n" + "="*50)
    print(f"ðŸš€ STARTING PIPELINE FOR: {short_name} ({model_id})")
    print("="*50)
    
    # A. Feature Extractor
    feature_extractor = AutoFeatureExtractor.from_pretrained(model_id)

    # B. Dataset Creation
    def preprocess_audio(batch, split="train"):
        audio_arrays = []
        for path in batch["audio_path"]:
            try:
                speech, sr = librosa.load(path, sr=Config.SAMPLING_RATE)
                if split == "train":
                    speech = augmentations(samples=speech, sample_rate=Config.SAMPLING_RATE)
                audio_arrays.append(speech)
            except:
                audio_arrays.append(np.zeros(Config.SAMPLING_RATE))
                
        inputs = feature_extractor(
            audio_arrays, 
            sampling_rate=Config.SAMPLING_RATE, 
            max_length=int(Config.SAMPLING_RATE * Config.MAX_DURATION), 
            truncation=True, 
            padding="max_length"
        )
        return {"input_values": inputs.input_values, "label": batch["label"]}

    train_ds = Dataset.from_pandas(train_df).map(lambda x: preprocess_audio(x, "train"), batched=True, batch_size=Config.BATCH_SIZE)
    eval_ds = Dataset.from_pandas(eval_df).map(lambda x: preprocess_audio(x, "test"), batched=True, batch_size=Config.BATCH_SIZE)
    test_ds = Dataset.from_pandas(test_df).map(lambda x: preprocess_audio(x, "test"), batched=True, batch_size=Config.BATCH_SIZE)

    # C. Load Model
    try:
        model = AutoModelForAudioClassification.from_pretrained(
            model_id,
            num_labels=Config.NUM_CLASSES,
            ignore_mismatched_sizes=True,
            attention_dropout=0.1,
            hidden_dropout=0.1,
            feat_proj_dropout=0.05
        ).to(Config.DEVICE)
        
        # FREEZE BACKBONE
        model.freeze_feature_encoder()
        
    except Exception as e:
        print(f"âŒ Failed to load {short_name}: {e}")
        continue

    # D. Training Config (Telugu Output Name)
    repo_name = f"{short_name}-telugu-binary"
    output_dir = repo_name
    
    if os.path.exists(output_dir): shutil.rmtree(output_dir)

    training_args = TrainingArguments(
        output_dir=output_dir,
        num_train_epochs=Config.EPOCHS,
        per_device_train_batch_size=Config.BATCH_SIZE,
        per_device_eval_batch_size=Config.BATCH_SIZE,
        gradient_accumulation_steps=Config.GRAD_ACCUMULATION,
        learning_rate=Config.LEARNING_RATE,
        weight_decay=0.01,
        warmup_ratio=0.1,
        logging_strategy="epoch",
        eval_strategy="epoch",
        save_strategy="epoch",
        load_best_model_at_end=True,
        metric_for_best_model="accuracy",
        save_total_limit=1,
        fp16=True if Config.DEVICE == "cuda" else False,
        report_to="none",
        push_to_hub=True,
        hub_model_id=f"{Config.HF_USERNAME}/{repo_name}",
        hub_token=Config.HF_TOKEN
    )

    # E. Initialize Trainer
    trainer = WeightedTrainer(
        model=model,
        args=training_args,
        train_dataset=train_ds,
        eval_dataset=eval_ds,
        processing_class=feature_extractor,
        data_collator=default_data_collator,
        compute_metrics=compute_metrics,
    )

    # F. Train & Save
    print(f"   ðŸƒâ€â™‚ï¸ Training {short_name}...")
    trainer.train()
    
    trainer.save_model(output_dir)
    
    # G. Evaluate
    print(f"   ðŸ“Š Evaluating {short_name} on Test Set...")
    test_results = trainer.predict(test_ds)
    print(f"   ðŸ† {short_name} Results:", test_results.metrics)

    # H. Push & Cleanup
    print(f"   â˜ï¸ Pushing {short_name} to Hub...")
    trainer.push_to_hub()
    
    print(f"   ðŸ§¹ Cleaning up {short_name}...")
    del model, trainer, train_ds, eval_ds, test_ds
    gc.collect()
    torch.cuda.empty_cache()
    if os.path.exists(output_dir): shutil.rmtree(output_dir)

print("\nâœ… ALL TELUGU AUDIO MODELS TRAINED SUCCESSFULLY!")
```

```python
# === MEMORY-SAFE META BUILD & TRAIN CELL (use existing eval_df & test_df) ===
# Requirements: transformers, torch, librosa, numpy, pandas, sklearn, lightgbm, joblib, tqdm
# pip install transformers librosa lightgbm joblib tqdm

import os, gc, sys, time
import numpy as np
import pandas as pd
from tqdm import tqdm
import torch, librosa
from transformers import (
    AutoTokenizer,
    AutoModelForSequenceClassification,
    AutoFeatureExtractor,
    AutoModelForAudioClassification,
)
import lightgbm as lgb
import joblib
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score

# -------- CONFIG (tune these if needed) ----------------
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
TEXT_BATCH_SIZE = 16          # batch size for text models (small text models OK)
AUDIO_BATCH_SIZE = 2          # very small to avoid OOM; set to 1 if still OOM
TARGET_SR = 16000
USE_DEVICE = DEVICE           # 'cuda' or 'cpu' (auto)
SAVE_PER_MODEL = False        # set True if you want per-model .npy saved

# -------- HF model repo paths (your best-epoch checkpoints) ----------
audio_model_paths = [
    "ramesh070/mms-telugu-binary",
    "ramesh070/wavlm-telugu-binary",
    "ramesh070/wav2vec2-telugu-binary",
]

text_model_paths = [
    "ramesh070/mbert-hatespeechdetection-telugu",
    "ramesh070/xlm-r-hatespeechdetection-telugu",
    "ramesh070/indicbert-hatespeechdetection-telugu",
]

# ---------------- helper utils --------------------------------
def softmax(x):
    e = np.exp(x - np.max(x, axis=-1, keepdims=True))
    return e / e.sum(axis=-1, keepdims=True)

def ensure_2class_probs(logits_or_probs):
    arr = np.asarray(logits_or_probs)
    if arr.ndim == 1:
        p1 = 1.0 / (1.0 + np.exp(-arr))
        p0 = 1.0 - p1
        return np.vstack([p0, p1]).T
    if arr.shape[1] == 2:
        s = arr.sum(axis=1)
        if not np.allclose(s, 1.0, atol=1e-3):
            return softmax(arr)
        return arr
    p = softmax(arr)
    if p.shape[1] >= 2:
        return p[:, :2]
    p1 = p.squeeze()
    p0 = 1.0 - p1
    return np.vstack([p0, p1]).T

def free_gpu():
    try:
        torch.cuda.empty_cache()
    except Exception:
        pass
    gc.collect()

# ---------------- text inference (one-model-at-a-time) -----------------------
def get_text_probs_from_model_streaming(model_name, texts, batch_size=TEXT_BATCH_SIZE, device=USE_DEVICE):
    """
    Load text model once, predict on texts in batches, then unload model.
    Returns (N,2) probabilities.
    Retries on CPU if OOM on GPU.
    """
    try_device = device
    for attempt in range(2):
        try:
            tok = AutoTokenizer.from_pretrained(model_name)
            model = AutoModelForSequenceClassification.from_pretrained(model_name)
            if try_device == "cuda" and torch.cuda.is_available():
                model.to("cuda")
            else:
                model.to("cpu")
            model.eval()

            probs_list = []
            with torch.no_grad():
                for i in range(0, len(texts), batch_size):
                    batch = texts[i:i+batch_size]
                    enc = tok(batch, truncation=True, padding=True, return_tensors="pt")
                    enc = {k: v.to(next(model.parameters()).device) for k, v in enc.items()}
                    out = model(**enc)
                    logits = out.logits.detach().cpu().numpy()
                    probs = ensure_2class_probs(logits)
                    probs_list.append(probs)

                    # free batch tensors
                    del enc, out, logits
                    free_gpu()

            # cleanup model
            model.to("cpu")
            del model, tok
            free_gpu()
            return np.vstack(probs_list)

        except RuntimeError as e:
            err = str(e)
            print(f"[text model {model_name}] RuntimeError: {err}")
            # if CUDA OOM, retry on CPU
            if 'out of memory' in err.lower() or try_device == "cuda":
                print(f"[text model {model_name}] retrying on CPU...")
                try_device = "cpu"
                free_gpu()
                continue
            else:
                raise e
    raise RuntimeError(f"Failed to run text model {model_name}")

# ---------------- audio inference (streaming, memory-safe) -------------------
def get_audio_probs_from_model_streaming(model_name, audio_paths, batch_size=AUDIO_BATCH_SIZE, target_sr=TARGET_SR, device=USE_DEVICE):
    """
    Memory-safe audio inference:
      - load model & feature extractor one at a time
      - stream audio files in tiny batches
      - if GPU OOM happens retry on CPU
    """
    try_device = device
    for attempt in range(2):
        try:
            feat = AutoFeatureExtractor.from_pretrained(model_name)
            model = AutoModelForAudioClassification.from_pretrained(model_name)
            if try_device == "cuda" and torch.cuda.is_available():
                model.to("cuda")
            else:
                model.to("cpu")
            model.eval()

            probs_list = []
            with torch.no_grad():
                for i in range(0, len(audio_paths), batch_size):
                    batch_paths = audio_paths[i:i+batch_size]
                    batch_wavs = []
                    for p in batch_paths:
                        try:
                            wav, sr = librosa.load(p, sr=target_sr)
                        except Exception:
                            # fallback 1s zeros if missing/corrupt
                            wav = np.zeros(int(target_sr * 1.0), dtype=np.float32)
                        batch_wavs.append(wav)

                    enc = feat(batch_wavs, sampling_rate=target_sr, return_tensors="pt", padding=True)
                    enc = {k: v.to(next(model.parameters()).device) for k, v in enc.items()}
                    out = model(**enc)
                    logits = out.logits.detach().cpu().numpy()
                    probs = ensure_2class_probs(logits)
                    probs_list.append(probs)

                    # free batch tensors
                    del enc, out, logits
                    free_gpu()

            # cleanup
            model.to("cpu")
            del model, feat
            free_gpu()
            return np.vstack(probs_list)

        except RuntimeError as e:
            err = str(e)
            print(f"[audio model {model_name}] RuntimeError: {err}")
            if 'out of memory' in err.lower() and try_device == "cuda":
                print(f"[audio model {model_name}] CUDA OOM - retrying on CPU (this will be slower).")
                try_device = "cpu"
                free_gpu()
                continue
            else:
                raise e
    raise RuntimeError(f"Failed to run audio model {model_name}")

def average_probs(list_of_probs):
    return np.mean(np.stack(list_of_probs, axis=0), axis=0)

# ------------------ ensure eval_df/test_df present -----------------------
if 'eval_df' not in globals() or 'test_df' not in globals():
    raise RuntimeError("eval_df/test_df not found in the notebook. Run the data-split cell first or load them into memory.")

# Basic sanity checks
for df_name in ('eval_df','test_df'):
    df = globals()[df_name]
    assert 'text' in df.columns and 'audio_path' in df.columns and 'label' in df.columns, \
        f"{df_name} must contain columns ['text','audio_path','label']"

print(f"DEVICE: {DEVICE} | TEXT_BATCH_SIZE: {TEXT_BATCH_SIZE} | AUDIO_BATCH_SIZE: {AUDIO_BATCH_SIZE}")
print(f"Eval size: {len(eval_df)} | Test size: {len(test_df)}")

# ---------------- produce per-model probs for eval ------------------------
print("Producing TEXT eval probs per model (one-at-a-time)...")
text_eval_probs_per_model = []
for m in text_model_paths:
    print("  ->", m)
    p = get_text_probs_from_model_streaming(m, eval_df['text'].tolist(), batch_size=TEXT_BATCH_SIZE, device=USE_DEVICE)
    text_eval_probs_per_model.append(p)
    if SAVE_PER_MODEL:
        np.save(f"text_eval_probs_{m.replace('/','_')}.npy", p)

text_eval_avg = average_probs(text_eval_probs_per_model)

print("Producing AUDIO eval probs per model (streaming, one-at-a-time)...")
audio_eval_probs_per_model = []
for m in audio_model_paths:
    print("  ->", m)
    p = get_audio_probs_from_model_streaming(m, eval_df['audio_path'].tolist(), batch_size=AUDIO_BATCH_SIZE, target_sr=TARGET_SR, device=USE_DEVICE)
    audio_eval_probs_per_model.append(p)
    if SAVE_PER_MODEL:
        np.save(f"audio_eval_probs_{m.replace('/','_')}.npy", p)

audio_eval_avg = average_probs(audio_eval_probs_per_model)

X_eval_meta = np.hstack([text_eval_avg, audio_eval_avg])  # shape (N_eval,4)
y_eval = eval_df['label'].astype(int).values

# ---------------- produce per-model probs for test ------------------------
print("Producing TEXT test probs per model...")
text_test_probs_per_model = []
for m in text_model_paths:
    print("  ->", m)
    p = get_text_probs_from_model_streaming(m, test_df['text'].tolist(), batch_size=TEXT_BATCH_SIZE, device=USE_DEVICE)
    text_test_probs_per_model.append(p)
    if SAVE_PER_MODEL:
        np.save(f"text_test_probs_{m.replace('/','_')}.npy", p)

text_test_avg = average_probs(text_test_probs_per_model)

print("Producing AUDIO test probs per model (streaming)...")
audio_test_probs_per_model = []
for m in audio_model_paths:
    print("  ->", m)
    p = get_audio_probs_from_model_streaming(m, test_df['audio_path'].tolist(), batch_size=AUDIO_BATCH_SIZE, target_sr=TARGET_SR, device=USE_DEVICE)
    audio_test_probs_per_model.append(p)
    if SAVE_PER_MODEL:
        np.save(f"audio_test_probs_{m.replace('/','_')}.npy", p)

audio_test_avg = average_probs(audio_test_probs_per_model)

X_test_meta = np.hstack([text_test_avg, audio_test_avg])  # (N_test,4)
y_test = test_df['label'].astype(int).values

# --------------- train LightGBM meta on eval (OOF) ------------------------
print("Training LightGBM meta on eval (OOF) features...")
params = {
    'objective': 'binary',
    'metric': 'binary_logloss',
    'num_leaves': 7,
    'max_depth': 3,
    'learning_rate': 0.05,
    'n_estimators': 150,
    'min_child_samples': 10,
    'verbosity': -1,
}
meta = lgb.LGBMClassifier(**params)
meta.fit(X_eval_meta, y_eval)

# Evaluate on test
y_prob_test = meta.predict_proba(X_test_meta)[:,1]
y_pred_test = (y_prob_test >= 0.5).astype(int)

print("Test Acc:", accuracy_score(y_test, y_pred_test))
print("Test F1 :", f1_score(y_test, y_pred_test))
try:
    print("Test AUC :", roc_auc_score(y_test, y_prob_test))
except Exception:
    pass

# Optional: find best threshold on eval (to maximize F1)
y_prob_eval = meta.predict_proba(X_eval_meta)[:,1]
best_t = 0.5; best_f1 = -1
for t in np.linspace(0.1, 0.9, 41):
    f = f1_score(y_eval, (y_prob_eval >= t).astype(int))
    if f > best_f1:
        best_f1 = f; best_t = t
print("Best threshold (tuned on eval):", best_t, "with F1:", best_f1)

# Save meta and feature arrays
joblib.dump(meta, "meta_lightgbm_tamil_from_eval_safe.joblib")
np.save("X_eval_meta.npy", X_eval_meta); np.save("y_eval.npy", y_eval)
np.save("X_test_meta.npy", X_test_meta); np.save("y_test.npy", y_test)
print("Saved meta and meta features. Done.")
```
